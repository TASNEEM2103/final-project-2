## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Get the Data

We will use a dataset for a cancer disease in the spine called kyphosis.
This dataset represents a number of patients who had kyphosis disease and had an operation.

Columns(Features) are:

Kyphosis: if the disease present or abscent after the operation 
Age: the age of patients in months 
Number: the no of vertebrate involved in the operation
Start: the no or the top most vertebrate that operated on

```python
data=pd.read_csv('kyphosis.csv') 
```
